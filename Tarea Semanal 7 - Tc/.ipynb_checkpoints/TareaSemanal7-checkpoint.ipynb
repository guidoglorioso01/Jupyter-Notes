```python
import numpy as np
from scipy.signal import butter, TransferFunction
from pytc2.sistemas_lineales import analyze_sys
import matplotlib.pyplot as plt
from scipy.signal import freqz,freqresp
# Parámetros del ejercicio 

fc = 1000.0  # Frecuencia de corte en Hz
orden = 2    # Orden del filtro
fs = 100e3   # -> frecuencia de muestreo para el filtro digital
fnyq = fs / 2
```

```python

##############################################################################
#### filtro analogico 
fc_normalizada = fc
num_a, den_a = butter(orden,fc_normalizada, btype='low', analog=True)
print("Coeficientes del filtro analogico normalizado:")
print("Coeficientes 'num':", num_a)
print("Coeficientes 'den':", den_a)

## Eje de frecuencia
frequencies = np.linspace(0, 10e3, 1000)  

#Respuesta del filtro
w_a, h_a = freqresp((num_a, den_a), w=frequencies)

modulo_fs_ideal = np.abs(h_a)
phase_fs_ideal = np.angle(h_a) 

## Ploteo señal obtenida

plt.figure(figsize=(12, 9))

# Gráfica de la magnitud
plt.subplot(2, 1, 1)
plt.title('Modulo del filtro')
plt.xlabel('Frecuencia (Hz)')
plt.ylabel('Magnitud (dB)')

plt.semilogx(w_a, 20 * np.log10(modulo_fs_ideal),label="Filtro Butter analogico")
plt.grid()
plt.legend()

# Gráfica de la fase
plt.subplot(2, 1, 2)
plt.title('Fase del filtro')
plt.xlabel('Frecuencia (Hz)')
plt.ylabel('Fase (radianes)')

plt.semilogx(w_a, phase_fs_ideal,label="Filtro Butter analogico")

plt.grid()

plt.tight_layout()
plt.legend()
plt.show()
```


En primer instancia calculo el filtro digital aplicando la transformada bilineal al filtro analogico. Para esto se resolvio en el PDF como queda la expresion del filtro.
Para poder ver la deformacion de una transformacion a la otra, se plotean tanto el butter analogico como el digital obtenido con la funcion bilineal.


```python
##############################################################################
##Aplicando bilineal obtuve los siguientes coeficientes (fs=100khz)
## Obtengo el filtro digital

K= 2*fs

A = den_a[1] # coeficientes de la transferencia sin normalizar
B = den_a[2]
C = num_a[0]

num_z_1 = [1*C,2*C,1*C] 
den_z_1 = [(K**2+A*K+B),(2*B-2*K**2),(K**2-A*K+B)]

w_1, h_1 = freqz(num_z_1, den_z_1, worN=frequencies, fs=fs)

modulo_fs1 = np.abs(h_1)
phase_fs1 = np.angle(h_1)

## Ploteo señal obtenida

plt.figure(figsize=(12, 9))

# Gráfica de la magnitud
plt.subplot(2, 1, 1)
plt.title('Modulo del filtro')
plt.xlabel('Frecuencia (Hz)')
plt.ylabel('Magnitud (dB)')

plt.semilogx(w_a, 20 * np.log10(modulo_fs_ideal),label="Filtro Butter analogico")
plt.semilogx(w_1, 20 * np.log10(modulo_fs1),label="Filtro digital FS = 100kHZ")
plt.grid()
plt.legend()

# Gráfica de la fase
plt.subplot(2, 1, 2)
plt.title('Fase del filtro')
plt.xlabel('Frecuencia (Hz)')
plt.ylabel('Fase (radianes)')

plt.semilogx(w_a, phase_fs_ideal,label="Filtro Butter analogico")
plt.semilogx(w_1, phase_fs1,label="Filtro digital FS = 100kHZ")
plt.grid()

plt.tight_layout()
plt.legend()
plt.show()
```

Ahora obtengo el mismo filtro anterior pero con una fs menor. En este caso al disminuir la fs se espera que sea mayor la distorcion de la transferencia respecto a la analogica
Nuevamente se compara con el filtro analogico para ver las diferencias.


```python
##############################################################################
## filtro con fs = 10khz
##
frequencies2 = np.linspace(0, 4e3, 1000) ## es necesario bajar la frecuencia maxima ya que cambio nyquist
fs= 10e3
K= 2*fs


A = den_a[1] # coeficientes de la transferencia sin normalizar
B = den_a[2]
C = num_a[0]

num_z_2 = [1*C,2*C,1*C] 
den_z_2 = [(K**2+A*K+B),(2*B-2*K**2),(K**2-A*K+B)]

w_2, h_2 = freqz(num_z_2, den_z_2, worN=frequencies2, fs=fs)

modulo_fs2 = np.abs(h_2)
phase_fs2 = np.angle(h_2)

## Ploteo señal obtenida

plt.figure(figsize=(12, 9))

# Gráfica de la magnitud
plt.subplot(2, 1, 1)
plt.title('Modulo del filtro')
plt.xlabel('Frecuencia (Hz)')
plt.ylabel('Magnitud (dB)')

plt.semilogx(w_a, 20 * np.log10(modulo_fs_ideal),label="Filtro Butter analogico")
plt.semilogx(w_2, 20 * np.log10(modulo_fs2),label="Filtro digital FS = 10kHZ")
plt.grid()
plt.legend()

# Gráfica de la fase
plt.subplot(2, 1, 2)
plt.title('Fase del filtro')
plt.xlabel('Frecuencia (Hz)')
plt.ylabel('Fase (radianes)')

plt.semilogx(w_a, phase_fs_ideal,label="Filtro Butter analogico")
plt.semilogx(w_2, phase_fs2,label="Filtro digital FS = 10kHZ")
plt.grid()

plt.tight_layout()
plt.legend()
plt.show()
```

```python


##############################################################################
## filtro con prewarp fs 100khz (no me estaria dando bien)
##
fs= 100e3
fc=1e3
K=fc/(2*np.tan(fc*np.pi/(2*fs))) # LA formula esta rara pero con esto da
print(f"El valor de K= {K}")

A = den_a[1] # coeficientes de la transferencia sin normalizar
B = den_a[2]
C = num_a[0]

num_z_3 = [1*C,2*C,1*C] 
den_z_3 = [(K**2+A*K+B),(2*B-2*K**2),(K**2-A*K+B)]

w_3, h_3 = freqz(num_z_3, den_z_3, worN=frequencies, fs=fs)

modulo_fs3 = np.abs(h_3)
phase_fs3 = np.angle(h_3)

## Ploteo señal obtenida

plt.figure(figsize=(12, 9))

# Gráfica de la magnitud
plt.subplot(2, 1, 1)
plt.title('Modulo del filtro')
plt.xlabel('Frecuencia (Hz)')
plt.ylabel('Magnitud (dB)')

plt.semilogx(w_a, 20 * np.log10(modulo_fs_ideal),label="Filtro Butter analogico")
plt.semilogx(w_3, 20 * np.log10(modulo_fs3),label="Filtro digital FS = 100kHZ con prewarp")
plt.grid()
plt.legend()

# Gráfica de la fase
plt.subplot(2, 1, 2)
plt.title('Fase del filtro')
plt.xlabel('Frecuencia (Hz)')
plt.ylabel('Fase (radianes)')

plt.semilogx(w_a, phase_fs_ideal,label="Filtro Butter analogico")
plt.semilogx(w_3, phase_fs3,label="Filtro digital FS = 100kHZ con prewarp")
plt.grid()

plt.tight_layout()
plt.legend()
plt.show()
```

Comparando todas las graficas tenemos los siguiente resultados:



```python

##############################################################################
## Crea las gráficas
plt.figure(figsize=(12, 9))

# Gráfica de la magnitud
plt.subplot(2, 1, 1)
plt.title('Modulo del filtro')
plt.xlabel('Frecuencia (Hz)')
plt.ylabel('Magnitud (dB)')


plt.semilogx(w_1, 20 * np.log10(modulo_fs1),label="Filtro digital FS = 100kHZ")
plt.semilogx(w_2, 20 * np.log10(modulo_fs2),label="Filtro digital FS = 10kHZ")
plt.semilogx(w_3, 20 * np.log10(modulo_fs3),label="Filtro digital FS = 100kHZ con prewarp")
plt.semilogx(w_a, 20 * np.log10(modulo_fs_ideal),label="Filtro Butter analogico",linestyle='dashed')

plt.grid()
plt.legend()

# Gráfica de la fase
plt.subplot(2, 1, 2)
plt.title('Fase del filtro')
plt.xlabel('Frecuencia (Hz)')
plt.ylabel('Fase (radianes)')

plt.semilogx(w_1, phase_fs1,label="Filtro digital FS = 100kHZ")
plt.semilogx(w_2, phase_fs2,label="Filtro digital FS = 10kHZ")
plt.semilogx(w_3, phase_fs3,label="Filtro digital FS = 100kHZ con prewarp")
plt.semilogx(w_a, phase_fs_ideal,label="Filtro Butter analogico",linestyle='dashed')

plt.grid()

plt.tight_layout()
plt.legend()
plt.show()
```

Con estas simulaciones podemos notar varias cosas y sacar conclusiones importantes:
Cambiar la frecuencia de muestreo implica que la respuesta del filtro sea mas o menos lineal para medias y altas frecuencias. Cuanto mayor es la frecuencia de muestreo
Podemos asegurar una mayor linealidad para frecuencias mas elevadas. Es por esto que en el caso de Fs=10Khz vemos la caida abrupta del modulo para frecuencias superiores a 1Khz mientras que para una fs de 100Khz esto no sucede.
Vemos ademas que modificar la frecuencia de muestreo NO desplaza al grafico en frecuencia. El modulo tiende a ser el mismo si no consideramos la alinealidad de las altas frecuencias.
Otro aspecto fundamental es tener en cuenta que nuestro filtro esta limitado por nyquist, es decir que bajo ningun concepto podemos utilizarlo para frecuencias mayores a la mitad de la de muestreo.
Como se observa tambien en las simulaciones al aplicar la transformada bilineal todos los modulos se desplazaron, dejando de cumplir con la plantilla del filtro analogico. 
Esto dependiendo de la implementacion puede ser un problema o no, y para solucionarlo, lo que se utiliza es el concepto de prewarp (señalizado con verde).
En este caso se aplico una norma de "distorcion" al mapeo de frecuencia anogica a la digital. Este factor que se aplico permite desplazar la respuesta para poder hacer conincidir un punto del filtro
analogico con un punto del filtro digital.

<h3>Ahora analizamos para el filtro pasa altos</h3>


```python

############################PASA ALTOS#######################################################

fc = 1000.0  # Frecuencia de corte en Hz
orden = 2    # Orden del filtro
fs = 100e3   # -> frecuencia de muestreo para el filtro digital
fnyq = fs / 2


### NOTA: a la hora de obtener el filtro analogico, si queremos normalizarlo es necesario hacerlo respecto de la frecuencia de nyquist para poder luego comparar con el digital

##############################################################################
#### filtro analogico 
fc_normalizada = fc
num_b, den_b = butter(orden,fc_normalizada, btype='high', analog=True)
print("Coeficientes del filtro analogico normalizado:")
print("Coeficientes 'num':", num_b)
print("Coeficientes 'den':", den_b)

## Eje de frecuencia
frequencies = np.linspace(10, 5e3, 1000)  # Rango de frecuencias de 0 a la mitad de fs

w_b, h_b = freqresp((num_b, den_b), w=frequencies)

modulo_fs_ideal = np.abs(h_b)
phase_fs_ideal = np.angle(h_b)


##############################################################################
##Aplicando bilineal obtuve los siguientes coeficientes (fs=100khz)
## Obtengo el filtro digital

K= 2*fs

A = den_b[1] # coeficientes de la transferencia sin normalizar
B = den_b[2]
C = num_b[0]

num_z_1 = [1*C*K**2,-2*C*K**2,1*C*K**2] 
den_z_1 = [(K**2+A*K+B),(2*B-2*K**2),(K**2-A*K+B)]

w_1, h_1 = freqz(num_z_1, den_z_1, worN=frequencies, fs=fs)

modulo_fs1 = np.abs(h_1)
phase_fs1 = np.angle(h_1)


##############################################################################
## filtro con fs = 10khz
##
fs= 10e3
K= 2*fs

A = den_b[1] # coeficientes de la transferencia sin normalizar
B = den_b[2]
C = num_b[0]

num_z_2 = [1*C*K**2,-2*C*K**2,1*C*K**2] 
den_z_2 = [(K**2+A*K+B),(2*B-2*K**2),(K**2-A*K+B)]

w_2, h_2 = freqz(num_z_2, den_z_2, worN=frequencies, fs=fs)

modulo_fs2 = np.abs(h_2)
phase_fs2 = np.angle(h_2)

##############################################################################
## filtro con prewarp fs 100khz 
##
fs= 100e3
fc=1e3
K=fc/(2*np.tan(fc*np.pi/(2*fs)))

A = den_b[1] # coeficientes de la transferencia sin normalizar
B = den_b[2]
C = num_b[0]

num_z_3 = [1*C*K**2,-2*C*K**2,1*C*K**2] 
den_z_3 = [(K**2+A*K+B),(2*B-2*K**2),(K**2-A*K+B)]

w_3, h_3 = freqz(num_z_3, den_z_3, worN=frequencies, fs=fs)

modulo_fs3 = np.abs(h_3)
phase_fs3 = np.angle(h_3)


##############################################################################
## Crea las gráficas
plt.figure(figsize=(12, 9))

# Gráfica de la magnitud
plt.subplot(2, 1, 1)
plt.title('Modulo del filtro')
plt.xlabel('Frecuencia (Hz)')
plt.ylabel('Magnitud (dB)')

plt.semilogx(w_1, 20 * np.log10(modulo_fs1),label="Filtro con Fs = 100Khz")
plt.semilogx(w_2, 20 * np.log10(modulo_fs2),label="Filtro con Fs = 10Khz")
plt.semilogx(w_3, 20 * np.log10(modulo_fs3),label="Filtro con Prewarp")
plt.semilogx(w_b, 20 * np.log10(modulo_fs_ideal),label="Filtro Butter analogico",linestyle='dashed')

plt.grid()
plt.legend()

# Gráfica de la fase
plt.subplot(2, 1, 2)
plt.title('Fase del filtro')
plt.xlabel('Frecuencia (Hz)')
plt.ylabel('Fase (radianes)')

plt.semilogx(w_1,phase_fs1,label="Filtro con Fs = 100Khz")
plt.semilogx(w_2,phase_fs2,label="Filtro con Fs = 10Khz")
plt.semilogx(w_3,phase_fs3,label="Filtro con Prewarp")
plt.semilogx(w_b,phase_fs_ideal,label="Filtro Butter analogico",linestyle='dashed')

plt.grid()

plt.tight_layout()
plt.legend()
plt.show()

```

En este caso vemos como aumentar la frecuencia de muestreo practicamente no tiene ningun cambio en la respuesta del filtro, sin embargo si vemos como el prewarp es una buena idea para cumplir con las condiciones de diseño del filtro.
