```python
from scipy.signal import TransferFunction,freqs
import matplotlib.pyplot as plt
import numpy as np
from pytc2.sistemas_lineales import analyze_sys

plt.close()

alpha_max= 1 #1db
alpha_min= 30 #30db
ws_pb = 4
wp_pb = 1 
epsilon = np.sqrt((10**(alpha_max/10))-1)
cuadrado = epsilon **2

print(f"Epsilon= {epsilon}")
for N in range(1,5):
    iteracion = 10* np.log10(1+epsilon**2 * ws_pb**(2*N) )
    print(f"para N={N} -> alpha_min = {iteracion}")
    
## 


def desnorm_trans(trans: TransferFunction ,norma: float):
    '''
    trans: transferencia que se desea desnormalizar o normalizar.
    norma: valor por el cual se quiere normalizar. 
    NOTA: $'.norma -> $ (si tenemos que desnormalizar una funcion debemos pasar 1/norma) 
    
    '''
    
    num_p, den_p = trans.num, trans.den
    
    num_p = [coeff*(norma**(len(num_p)-1-i)) for i,coeff in enumerate(num_p)] # aplico la normalizacion o desnormalizacion que queramos a la tf
    den_p = [coeff*(norma**(len(den_p)-1-i)) for i,coeff in enumerate(den_p)]
    num_p = np.multiply(num_p , (norma)**-3)
    den_p = np.multiply(den_p , (norma)**-3)

    tf_completa = TransferFunction(num_p, den_p)
    
    return tf_completa
```

A continuacion se simula el filtro ya convertido mediante el nucleo de transformacion obteniendo asi el filtro pasa altos normalizado.
Se puede notar como ahora la normalizacion de butter desplazara el grafico en el otro sentido ya que al 

```python

##simulacion del filtro convertido a pasa altos con el nucleo de transformacion $ -> 1/$'

## Transferencia segundo orden

tf_segundo_den = [1, 0, 0]
tf_segundo_num =  [1, 1 , 1]

## Transferencia primer orden

tf_primero_den = [1,0]
tf_primero_num =  [1, 1]

## Transferencia Sistema completo

tf_completa = TransferFunction(np.polymul(tf_primero_den, tf_segundo_den),np.polymul(tf_segundo_num,tf_primero_num))

## Simulacion sin normas de frecuencia
analyze_sys( tf_completa, "Pasa altos normalizado")

##simulacion sin omega de butter
ep = 1/1.2525 # debo invertirlo ya que pasa por el nucleo de transformacion 
tf_completa = desnorm_trans(tf_completa ,norma = 1/ep )

analyze_sys( tf_completa, "Pasa altos con wb")

```

```python
## Simulacion del filtro pasa bajos (plantilla) normalizado con omega de butter

## Transferencia segundo orden

tf_segundo_den = [1]
tf_segundo_num =  [1, 1 , 1]

## Transferencia primer orden

tf_primero_den = [1]
tf_primero_num =  [1, 1]

## Transferencia Sistema completo

tf_completa = TransferFunction(np.polymul(tf_primero_den, tf_segundo_den),np.polymul(tf_segundo_num,tf_primero_num))


## Simulacion Normalizada

analyze_sys( tf_completa, "Pasa bajos norm" )

##simulacion sin omega de butter
ep = 1.2525 
tf_completa = desnorm_trans(tf_completa ,norma = 1/ep )

analyze_sys( tf_completa, "Pasa bajos con wb" )

```


A la hora de implementarlo se observo que la inductancia necesaria era de 5uHY, haciendo simulaciones con los valores correspondientes se observa que no se pueden conseguir estos valores.Lo que ocurre en este caso es que la capacidad colocada es tan baja que la capacidad interna de los operacionales entra en juego y no permite conseguir la inductancia requerida.
El problema de seguir subiendo la capacidad y disminuyendo la resistencia es que ahora el GIC tiene valores tan bajos que el consumo de corriente que requiere no es viable, es por ello que se aplica una norma de impedancia a todo el circuito con el fin de aumentar la inductancia y en concecuencia el valor de los componentes
Es por ello que la norma de impedancia aca juega un rol fundamental.
Como conclusion del GIC es conveniente no usar valores de capacitores por debajo de los 10nf, ya que el efecto de los operacionales empieza a ser considerable en la transferencia.
