A continuacion se muestra una simulacion como es la respuesta de un filtro chebyshef para distintos n


```python
# -*- coding: utf-8 -*-
"""
Created on Sun Apr 30 18:26:39 2023

@author: guido

Grafica de transferencia de chebyshef para distintos n
"""


import matplotlib.pyplot as plt
import numpy as np

# Definimos el rango de valores de x que queremos graficar
w = np.arange(0, 2, 0.01)

## funcion que quiero plotear
epsilon = 1
n=(2,5) ## Aca se puede fijar todos los n que queramos comparar

for this_n in n:
    cn_mayor1= epsilon * np.cosh(this_n*np.arccosh(w))
    cn_menor1= epsilon * np.cos(this_n*np.arccos(w))
    
    y = np.where(w >= 1, 1/(1+cn_mayor1**2) , 1/(1+cn_menor1**2) )
    
    # Creamos la gráfica con los valores de x e y
    plt.plot(w, y)

    
plt.xlim(0, 2)
plt.ylim(0, 1.1)
  
# Agregamos una línea punteada en y=1 y en y= 1/(1+Epsilon)
plt.axhline(y=1, color='gray', linestyle='--')
plt.axhline(y=1/(1+epsilon), color='gray', linestyle='--')
    
    
# Agregamos etiquetas a los ejes y un título a la gráfica
plt.xlabel('w')
plt.ylabel('|T(jw)|^2')
plt.title('Respuesta de chebyshef')


```
