Lo que se puede observar para el caso de S11 es como a medida que aumenta la frecuencia dicha funcion crece. Esto tiene mucho sentido ya que S11 es un coeficiente de reflexion, por lo que es de esperar q a medida que el filtro atenua mas, esa potencia es reflejada hacia el generador. Por ultimo vemos que para f-> inf lo que vemos es un coefienciete de reflexion = 1 por lo que la atenuacion es total y no se transmite potencia a la carga.
