```python
import numpy as np
from scipy import signal as signal
from pytc2.sistemas_lineales import analyze_sys, pretty_print_bicuad_omegayq,tfcascade,pretty_print_lti
```

Como se puede observar en la resolucion el orden del filtro tiene que ser de N=3. Veamos por que esto debe ser asi.
En primer instancia vemos que la transferencia tiene un punto donde la transferencia tiende a menos infinito. Esto es una propiedad que se obtiene al colocar un par de ceros complejos conjugados, ya que cuando la omega este sobre el cero la transferencia tiende a cero, por lo que en db esto se ve como una atenuacion de infinitos dB.
A continuacion se muestre este efecto

```python
#representacion de un par de ceros hubicados en el eje jw en +-3j
num = [1/9, 0,1]
den = [1]
H1 = signal.TransferFunction(num,den)
pretty_print_lti(num,den)
analyze_sys(H1)
```

Como se puede observar en la consigna el filtro es del tipo butter, pues es maxima planicidad y para la omega normalizada tiene -3dB. Es por ello que sabemos que el filtro tendra que ser necesariamente del tipo Butter.
A continuacion se representa como quedaria la transferencia si fuera un butter de orden 2:

```python
## Grafica de N = 2

N = 2
wp = 1
num_but, den = signal.butter(N,  wp, btype='low', analog=True, output='ba') 
num_but = num # Dejo el par de ceros en la transferencia
H1 = signal.TransferFunction(num_but,den)
pretty_print_lti(num_but,den)
analyze_sys(H1,"PB N=2 con ceros")
```

Como se pudo observar el filtro una vez pasado el cero no sigue atenuando. Esto se debe a que el par de ceros contrarrestra la atenuacion de los polos, y al ser del mismo orden ambos la respuesta del modulo se mantedra constante. Por lo tanto al ver que en la consigna atenua 20db por decada luego de los ceros es necesario agregar un polo mas que permita mantener esa atenuacion luego de pasar por los ceros.

```python
## Transferencia con el N = 3

N = 3
wp = 1
num_but, den = signal.butter(N,  wp, btype='low', analog=True, output='ba')
num_but = num ## conservo los ceros
H1 = signal.TransferFunction(num_but,den)
pretty_print_lti(num_but,den)
analyze_sys(H1,"PB N=3 con ceros")
```

Si analizamos en detalle la grafica obtenida podemos observar los siguientes puntos:



Donde se ve que para la frecuencia de corte tenemos 4dB en vez de 3dB, esto se debe a que el cero esta provocando que la transferencia atenue mas rapido al rededor del punto donde se coloco.



Aca se puede observar como la transferencia una vez pasado el cero tiene una atenuacion de 20dB / dec al igual que en la plantilla pedida.

Por ultimo podemos poner ambas transferencias juntas para ver su similitud.






Ahora se procede a aplicar el nucleo de transformacion para lo cual por medio de la transferencia obtenida remplazamos $ -> 1/$'



obtenemos:



```python
den = [1,2,2,1]
num = [1,0,1/9,0]

pretty_print_lti(num,den)
```

```python
# Podriamos haber obtenido la misma transferencia utilizando las funciones de butter para filtro pasa altos, y convirtiendo el cero del pasa bajos por medio del nucleo de 
#transformacion a una transferencia pasa altos. Por ultimo multiplicar ambas transferencias
N= 3
wp = 1
num_but, den_but = signal.butter(N,  wp, btype="high", analog=True, output='ba')

print("Transferencia pasa altos de N = 3")
pretty_print_lti(num_but,den_but)

print("Antes de aplicar el nucleo a los ceros")
num = [1/9,0,1] # Antes de aplicar nucleo de transformacion
pretty_print_lti(num,1)

print("Luego de aplicar el nucleo a los ceros")
num_cero = [1,0,1/9] # luego de aplicar nucleo
den_cero = [1,0,0]
pretty_print_lti(num_cero,den_cero)

print("multiplicando ambas transferencias:")
den = [1,2,2,1]
num = [1,0,1/9,0]
pretty_print_lti(num,den)

print("Transferencia del filtro:")
H1 = signal.TransferFunction(num,den)
analyze_sys(H1,"Transferencia HP")
```

Circuito utilizado:





En este circuito, se utilizan dos etapas diferentes. La primera etapa es un filtro pasa altos de primer orden, implementado mediante un circuito R-C. Se incluye un amplificador operacional para evitar la carga de la etapa siguiente.

La segunda etapa es un circuito propuesto que consta de un Generador de Impedancia Controlada (GIC) que actúa como un girador para simular una bobina. En combinación con un circuito R-C, forma un filtro de segundo orden. Esta configuración permite tener una mayor flexibilidad en la implementación del filtro y lograr una respuesta más precisa.

Además, el circuito cuenta con una pre-inyección a través de la resistencia R5. Esto permite modificar la transferencia del circuito y, a su vez, permite la inserción de ceros en la respuesta de frecuencia. Esta característica es especialmente útil, ya que permite colocar dos polos y dos ceros en la transferencia, brindando un mayor control y ajuste de la respuesta del filtro.

En resumen, este circuito resulta ideal para implementaciones en las que se requiera una configuración con múltiples polos y ceros, además de ofrecer la ventaja de tener una salida proporcionada por el amplificador operacional.
